```python
import numpy as np
import matplotlib.pyplot as plt

def make_spiral(n_point=300, n_classes=3, noise=0.5):

    X = np.zeros((n_point * n_classes, 2))
    y = np.zeros(n_point * n_classes, dtype='int')

    for class_number in range(n_classes):

        ix = range(
            n_point * class_number,
            n_point * (class_number + 1)
        )

        r = np.linspace(0.0, 1, n_point)

        t = np.linspace(
            class_number * 4,
            (class_number + 1) * 4,
            n_point
        ) + np.random.randn(n_point) * noise

        X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]

        y[ix] = class_number

    return X, y


X, y = make_spiral(
    n_point=300,
    n_classes=3,
    noise=0.5
)

plt.figure(figsize=(8,8))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    cmap=plt.cm.Spectral,
    s=28,
    edgecolors='k'
)

plt.title("2D Spiral Data with Noise")

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
import numpy as np

class NN:

    def __init__(self, n_features, n_hidden, n_classes):

        self.d = n_features
        self.h = n_hidden
        self.n = n_classes

        # weights and biases
        self.W1 = 0.01 * np.random.randn(self.d, self.h)
        self.b1 = np.zeros((1, self.h))

        self.W2 = 0.01 * np.random.randn(self.h, self.n)
        self.b2 = np.zeros((1, self.n))

    # Forward propagation
    def fwd_props(self, X):

        # Hidden layer
        Z1 = np.dot(X, self.W1) + self.b1

        # ReLU activation
        A1 = np.maximum(0, Z1)

        # Output layer
        Z2 = np.dot(A1, self.W2) + self.b2

        # Softmax
        exp_scores = np.exp(Z2)

        A2 = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

        return A1, A2

    # Compute loss
    def c_loss(self, y, probs):

        num_examples = y.shape[0]

        correct_logprobs = -np.log(probs[range(num_examples), y])

        loss = np.sum(correct_logprobs) / num_examples

        return loss

    # Backpropagation
    def back_prob(self, X, A1, A2, y):

        num_examples = X.shape[0]

        dZ2 = A2.copy()

        dZ2[range(num_examples), y] -= 1

        dZ2 /= num_examples

        # Gradients for W2 and b2
        dW2 = np.dot(A1.T, dZ2)

        db2 = np.sum(dZ2, axis=0, keepdims=True)

        # Backprop into hidden layer
        dA1 = np.dot(dZ2, self.W2.T)

        # ReLU derivative
        dA1[A1 <= 0] = 0

        dZ1 = dA1

        # Gradients for W1 and b1
        dW1 = np.dot(X.T, dZ1)

        db1 = np.sum(dZ1, axis=0, keepdims=True)

        return dW1, db1, dW2, db2

    # Training
    def fit(self, X, Y, lr, reg, max_iters):

        for i in range(max_iters):

            # Forward
            A1, A2 = self.fwd_props(X)

            # Loss
            data_loss = self.c_loss(Y, A2)

            reg_loss = (
                0.5 * reg * np.sum(self.W1 * self.W1)
                + 0.5 * reg * np.sum(self.W2 * self.W2)
            )

            loss = data_loss + reg_loss

            if i % 1000 == 0:
                print("Iteration %d: loss %f" % (i, loss))

            # Backprop
            dW1, db1, dW2, db2 = self.back_prob(X, A1, A2, Y)

            # Regularization gradient
            dW2 += reg * self.W2
            dW1 += reg * self.W1

            # Gradient descent update
            self.W1 -= lr * dW1
            self.b1 -= lr * db1

            self.W2 -= lr * dW2
            self.b2 -= lr * db2

    # Prediction
    def predict(self, X):

        A1 = np.maximum(0, np.dot(X, self.W1) + self.b1)

        Z2 = np.dot(A1, self.W2) + self.b2

        y_hat = np.argmax(Z2, axis=1)

        return y_hat








    
```

```python
nn_model = NN(
    n_features=2,
    n_hidden=100,
    n_classes=3
)

nn_model.fit(X, y, lr=1, reg=1e-3, max_iters=10000)
```

```python
Z = nn_model.predict(X)
unique_values  , counts = np.unique(Z , return_counts=True)

print("Unique Values:" , unique_values)
print("Counts:" , counts)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step size
h = 0.01

# Create mesh grid limits
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# Mesh grid
xx, yy = np.meshgrid(
    np.arange(x_min, x_max, h),
    np.arange(y_min, y_max, h)
)

# Flatten grid points
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Predict for every point
Z = nn_model.predict(grid_points)

# Reshape predictions
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(10, 8))

plt.contourf(
    xx,
    yy,
    Z,
    alpha=0.8,
    cmap=plt.cm.Spectral
)

# Plot training points
plt.scatter(
    X[:, 0],
    X[:, 1],
    c=y,
    s=30,
    cmap=plt.cm.Spectral,
    edgecolors='k'
)

# Labels
plt.title("Neural Network Decision Boundary")

plt.xlabel("X1")
plt.ylabel("X2")

plt.show()

```
