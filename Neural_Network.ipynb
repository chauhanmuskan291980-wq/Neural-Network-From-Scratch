```python
import numpy as np
import matplotlib.pyplot as plt

N = 500

theta = np.linspace(0, 4*np.pi, N)

x = theta * np.cos(theta)
y = theta * np.sin(theta)

# Add noise
x += np.random.randn(N) * 0.5
y += np.random.randn(N) * 0.5

X = np.column_stack((x, y))

Y = theta

plt.figure(figsize=(8,8))

plt.scatter(
    X[:,0],
    X[:,1],
    c=Y,
    cmap=plt.cm.Spectral
)

plt.title("2D Spiral Dataset")

plt.xlabel("X")
plt.ylabel("Y")

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def make_spiral(n_point=300, n_classes=3, noise=0.5):

    X = np.zeros((n_point * n_classes, 2))
    y = np.zeros(n_point * n_classes, dtype='int')

    for class_number in range(n_classes):

        ix = range(
            n_point * class_number,
            n_point * (class_number + 1)
        )

        r = np.linspace(0.0, 1, n_point)

        t = np.linspace(
            class_number * 4,
            (class_number + 1) * 4,
            n_point
        ) + np.random.randn(n_point) * noise

        X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]

        y[ix] = class_number

    return X, y


X, y = make_spiral(
    n_point=300,
    n_classes=3,
    noise=0.5
)

plt.figure(figsize=(8,8))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y,
    cmap=plt.cm.Spectral,
    s=28,
    edgecolors='k'
)

plt.title("2D Spiral Data with Noise")

plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```
